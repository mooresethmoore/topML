```python
import pandas as pd
import ast, itertools, os, io,pickle
dirname = "."
#df = pd.read_csv(os.path.join(dirname, "train_pdStack_index_b4_5.pkl.csv"), index_col=0)
dfs=[pd.read_csv(os.path.join(dirname, f"train_pdStack_index_b{j}_5.pkl.csv"), index_col=0) for j in range(5)]
df = pd.concat(dfs,ignore_index=True,axis=0)
df["linkers_cansmi"] = [ast.literal_eval(df["linkers_cansmi"].to_list()[x]) for x in df.index]
total_linkers = list(itertools.chain.from_iterable(df["linkers_cansmi"].to_list()))
unique_linkers = list(set(total_linkers))
df["nodes_cansmi"] = [ast.literal_eval(df["nodes_cansmi"].to_list()[x]) for x in df.index]
total_nodes = list(itertools.chain.from_iterable(df["nodes_cansmi"].to_list()))
unique_nodes = list(set(total_nodes))
df
```

```python
import matplotlib.pyplot as plt
linker_index_map = dict(zip(list(set(total_linkers)), ["L"+str(x) for x in range(len(list(set(total_linkers))))]))
total_linkers_idx = pd.Series(total_linkers).map(linker_index_map).to_list()
import collections
histog = collections.Counter(total_linkers_idx)
linker_df = pd.DataFrame([histog.keys(), histog.values()]).T
linker_df.columns = ["linker_id", "freq"]
linker_df["smi"] = linker_df["linker_id"].map({v: k for k, v in linker_index_map.items()})
linker_df = linker_df[linker_df["smi"]!="[CH3]"]
linker_df = linker_df[linker_df["smi"]!="N#N"]
linker_df = linker_df[(linker_df["smi"].str.contains("C"))|(linker_df["smi"].str.contains("c"))]
linker_df = linker_df.sort_values(by="freq", ascending=False)
linker_df = linker_df[linker_df["freq"]>100].reset_index(drop=True)
_df = df.dropna()

fig, ax = plt.subplots(figsize=(20, 10))
ax.bar(linker_df["linker_id"], linker_df["freq"], width=0.3)
for i in linker_df.index:
    ax.text(i-0.3, 6, linker_df.at[i, "smi"], rotation=90)
    ax.text(i-0.1, linker_df.at[i, "freq"] + 10, "%.2f" % (linker_df.at[i, "freq"] * 100. / len(_df)) + "%")
ax.set_ylabel("Frequency")
ax.set_xlabel("Linker IDs")
ax.set_title("Histogram of Top Linkers")
#fig.savefig(os.path.join(dirname, "linker_hist.png"), dpi=300, transparent=True)


from openbabel import pybel
xyzstrs = []
for x in linker_df.index:
    mol = pybel.readstring("smi", linker_df.at[x, "smi"].split()[0].strip())
    mol.make3D()
    print(linker_df.at[x, "smi"])
    display(mol)
    xyzstrs.append(mol.write("xyz"))
    pngname = os.path.join(dirname, linker_df.at[x, "linker_id"]+".png")
    mol.draw(show=False, filename=pngname)
    with io.open(pngname.replace(".png", ".xyz"), "w", newline="\n") as wf:
        wf.write(xyzstrs[-1])

from PIL import Image

img1 = Image.open(os.path.join(dirname, "linker_hist.png")).convert("RGBA")
newImage = Image.new("RGBA", (img1.width, img1.height+200))
newImage.paste(img1, (0,0), mask=img1)

for i in linker_df.index:
    img2 = Image.open(os.path.join(dirname, linker_df.at[i, "linker_id"] + ".png")).convert("RGBA")
    newImage.paste(img2, (800 + (i * img2.width), 2800), mask=img2)
newImage.show()
newImage.save(os.path.join(dirname, "linker_hist_mol.png"))
```

```python
import matplotlib.pyplot as plt
node_index_map = dict(zip(list(set(total_nodes)), ["L"+str(x) for x in range(len(list(set(total_nodes))))]))
total_nodes_idx = pd.Series(total_nodes).map(node_index_map).to_list()
import collections
histog = collections.Counter(total_nodes_idx)
node_df = pd.DataFrame([histog.keys(), histog.values()]).T
node_df.columns = ["node_id", "freq"]
node_df["smi"] = node_df["node_id"].map({v: k for k, v in node_index_map.items()})
node_df = node_df.sort_values(by="freq", ascending=False)
node_df = node_df[node_df["freq"]>10].reset_index(drop=True)
fig, ax = plt.subplots()
ax.bar(node_df["node_id"], node_df["freq"], width=0.3)
for i in node_df.index:
    text = node_df.at[i, "smi"]
    dx = 0
    sep_len = 41
    if len(node_df.at[i, "smi"]) > sep_len:
        text = node_df.at[i, "smi"][:sep_len] + "\n" + node_df.at[i, "smi"][sep_len:]
        dx = dx + 0.15
    ax.text(i-0.3 - dx, 100, text, rotation=90)
ax.set_ylabel("Frequency")
ax.set_xlabel("Node IDs")
ax.set_title("Histogram of Top Nodes")
fig.savefig(os.path.join(dirname, "node_hist.png"), dpi=300, transparent=True)

from openbabel import pybel
xyzstrs = []
for x in node_df.index:
    mol = pybel.readstring("smi", node_df.at[x, "smi"].split()[0].strip())
    mol.make3D()
    print(node_df.at[x, "smi"])
    display(mol)
    xyzstrs.append(mol.write("xyz"))
    pngname = os.path.join(dirname, node_df.at[x, "node_id"]+".png")
    mol.draw(show=False, filename=pngname)
    with io.open(pngname.replace(".png", ".xyz"), "w", newline="\n") as wf:
        wf.write(xyzstrs[-1])


from PIL import Image

img1 = Image.open(os.path.join(dirname, "node_hist.png")).convert("RGBA")
newImage = Image.new("RGBA", (img1.width, img1.height+200))


for i in node_df.index:
    img2 = Image.open(os.path.join(dirname, node_df.at[i, "node_id"] + ".png")).convert("RGBA")
    newImage.paste(img2, (220 + (i * img2.width), img1.height-80), mask=img2)
newImage.paste(img1, (0,0), mask=img1)
newImage.show()
newImage.save(os.path.join(dirname, "node_hist_mol.png"))
```

```python
linker_belong_dict = dict(zip(linker_df["linker_id"].to_list(), [[] for i in linker_df.index]))
node_belong_dict = dict(zip(node_df["node_id"].to_list(), [[] for i in node_df.index]))
for i in df.index:
    for l in df.at[i, "linker_ids"]:
        if l in linker_belong_dict.keys():
            linker_belong_dict[l].append(df.at[i, "cif"])
    for n in df.at[i, "node_ids"]:
        if n in node_belong_dict.keys():
            node_belong_dict[n].append(df.at[i, "cif"])
            
import json
 
# Writing to sample.json
with io.open(os.path.join(dirname, "linker_belong_dict.json"), "w", newline="\n") as outfile:
    outfile.write(json.dumps(linker_belong_dict, indent=4))
with io.open(os.path.join(dirname, "node_belong_dict.json"), "w", newline="\n") as outfile:
    outfile.write(json.dumps(node_belong_dict, indent=4))
```
